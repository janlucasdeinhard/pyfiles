```python
import numpy as np 
from numpy import random as rnd 
from matplotlib import pyplot as plt 
import warnings,datetime,time,os,sys,math,itertools,platform  

import pandas as pd 
```

```python
if platform.system()=='Windows':
    filepath = '../../TF_Data/Dropbox/PhD Prep/DQ Framework - Clustering/Ontological Description of Christmas 2023 Issues/Recent MWF Issues Ontological Description.txt'
else:
    raise Exception('Please configure the system path on the current platform!')
```

```python
txt = []
with open(filepath) as f:
    l = f.readline()
    while l:
        txt.append(l)
        l = f.readline()
```

```python
df = pd.DataFrame([k.replace('\n','').split(' - ') for k in txt[txt.index('Derived triple structure:\n'):] if (len(k.replace('\n',''))>0 and k!='Derived triple structure:\n')]).rename(columns={
    0:'Subject',
    1:'Predicate',
    2:'Object'
}).drop_duplicates()
```

```python
Class_list = list(set(df['Subject'].tolist() + df['Object'].tolist()))
```
