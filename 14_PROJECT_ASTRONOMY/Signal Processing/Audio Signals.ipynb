```python
import numpy as np
from numpy import random as rnd
from matplotlib import pyplot as plt
import os,sys,warnings,datetime,time,math,itertools,platform

from scipy.io import wavfile
from scipy.fft import fft,fftfreq

import simpleaudio as sa
```

```python
# Data path depends on machine
datapath_windows = r'../../../TF_Data/Dropbox/TF_data/Space Sounds/'

# Select relevant datapath
if platform.system()=='Windows':
    filepath = datapath_windows
filename = 'star_KIC12268220C_kepler_LightToSoundCurve.wav'
```

```python
sr,data = wavfile.read(filepath+filename)

wav_file = sa.WaveObject.from_wave_file(filepath+filename)
```

```python
t = np.arange(data.shape[0])/sr

freqs = fftfreq(data.shape[0],sr)
X = fft(data)
```

```python
play_obj = wav_file.play()

fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(30,10))

ax[0].plot(t,data)
ax[0].set_xlabel('Time t [s]')
ax[0].set_ylabel('Signal $x(t)$')
ax[0].set_title('RX Signal over time')
ax[0].grid()

ax[1].plot(freqs,np.abs(X))
ax[1].set_xlabel('Frequencies $\omega$ [Hz]')
ax[1].set_ylabel('Frequency Decomposition $X(\omega)$')
ax[1].set_title('Frequency Decomposition')
ax[1].set_xlim([-3e-7,3e-7])
ax[1].grid()

plt.show()
```

# Audio Recording
This audio recording was sourced from the transformation of the recorded light curve of star KIC 12268220C into the audio domain. The Fourier transform is limited to the interval $[-3\times 10^{-7},3\times 10^{-7}]\text{ Hz}$. the frequencies of this acis translate into a recurrence of about $\frac{10^{7}\text{ s}}{3600\text{ s per hour}\times 24\text{ hours per day}}\approx 116\text{ days}$.
