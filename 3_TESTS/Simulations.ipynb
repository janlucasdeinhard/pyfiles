```python
# Imports
import numpy as np
from numpy import random as rnd
from matplotlib import pyplot as plt
from scipy.stats import gaussian_kde

# Funtion to simulate N times a group of a given group size and 
# extract the maximum IQ of that group as an approximation to group IQ  
def sample_coop(coop_size,N=10000):
    # List to store coop IQ samples
    x_coop = []
    # Iterate N samples
    for n in range(N): x_coop.append(np.max(100 + 20*rnd.randn(coop_size)))
    # Convert to array
    x_coop = np.array(x_coop)
    # Return result
    return x_coop

# Simulate IQ distributions for different group sizes
x1 = gaussian_kde(sample_coop(1))
x10 = gaussian_kde(sample_coop(10))
x100 = gaussian_kde(sample_coop(100))
x1000 = gaussian_kde(sample_coop(1000))
# Smoothen distributions
x1.covariance_factor = lambda: .5
x1._compute_covariance()
x10.covariance_factor = lambda: .5
x10._compute_covariance()
x100.covariance_factor = lambda: .5
x100._compute_covariance()
x1000.covariance_factor = lambda: .5
x1000._compute_covariance()
# IQ axis for the plots
x = np.linspace(0,300,10000)
# Plot
plt.plot(x,x1(x))
plt.plot(x,x10(x))
plt.plot(x,x100(x))
plt.plot(x,x1000(x))
# Generate legend
plt.legend(['Group size 1','Group size 10','Group size 100','Group size 1000'])
# Generate output
plt.show()
```
