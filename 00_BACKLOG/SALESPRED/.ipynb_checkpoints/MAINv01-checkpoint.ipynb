```python
import numpy as np
from numpy import random as rnd
from matplotlib import pyplot as plt
import datetime,os,warnings,csv

import tensorflow as tf
import pandas as pd
import seaborn as sns

from keras.models import Model,Sequential
from keras.layers import Input,Dense,LSTM,advanced_activations,Dropout,BatchNormalization,concatenate,Embedding

from sklearn.model_selection import train_test_split
from sklearn import preprocessing as pp
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor

sns.set()
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('..\\..\\TF_data\\SALESPRED\\walmart\\train.csv',infer_datetime_format=True)
df['IsHoliday'] = df['IsHoliday'].apply(lambda x: int(x))
df['Date'] = pd.to_datetime(df['Date'],format='%Y-%m-%d')
```

```python
def generator(df,lookback=100,forecast=1,batch_size=3):
    
    num_stores = len(list(set(df['Store'])))
    num_depts = len(list(set(df['Dept'])))
    
    while True:
        
        cstore = rnd.randint(1,num_stores) 
        cdept = rnd.randint(1,num_depts)
        
        salesdata = df[(df['Store']==cstore)&(df['Dept']==cdept)].sort_values(by=['Date'],ascending=True)['Weekly_Sales'].values
        for k in range(lookback,len(salesdata)-forecast):
            
            try:
                X = np.zeros(shape=(batch_size,lookback))
                y = np.zeros(shape=(batch_size,forecast))
                for kk in range(batch_size):
                    X[kk,:] = salesdata[k-lookback:k]
                    y[kk,:] = salesdata[k:k+forecast]
            except:
                break
            yield X,y
```

```python
g = generator(df)
```

```python
lookback = 100 
forecast = 5

additional_information = []

timeseries = []
d_timeseries = []

labels = []
labels_d = []

storeIDrange = max(df['Store'].unique())
for storeID in df['Store'].unique():
    for deptID in df['Dept'].unique():
        t = df[(df['Store']==storeID)&(df['Dept']==deptID)].sort_values(by='Date')['Weekly_Sales'].values
        #dt = df[(df['Store']==storeID)&(df['Dept']==deptID)].sort_values(by='Date')['Weekly_Sales'].diff().values
        info = [storeID,deptID]
        
        print('Processing store %d/%d, department %d'%(storeID,storeIDrange,deptID))
        for k in range(lookback,t.shape[0]-forecast):
            additional_information.append(info)
            timeseries.append(t[k-lookback:k])
            #d_timeseries.append(dt[k-lookback:k])
            labels.append(t[k+forecast]-t[k])
            #labels_d.append(dt[k+forecast]-dt[k])
                    
additional_information = np.stack(additional_information,axis=0)
timeseries = np.expand_dims(np.stack(timeseries,axis=0),axis=2)
#d_timeseries = np.expand_dims(np.stack(d_timeseries,axis=0),axis=2)
labels = np.stack(labels,axis=0)
#labels_d = np.stack(labels_d,axis=0)

Xtrain,Xtest,Ztrain,Ztest,ytrain,ytest = train_test_split(timeseries,additional_information,labels,test_size=0.5,shuffle=True)

ytrain = np.expand_dims(ytrain,axis=1)
ytest = np.expand_dims(ytest,axis=1)

disc = pp.KBinsDiscretizer(n_bins=8)
ytrain_oh = disc.fit_transform(ytrain).todense()
ytest_oh = disc.transform(ytest).todense()
```

```python
act_fun = 'tanh'

in0 = Input(shape=(lookback,1,))
x = LSTM(units=100,activation=act_fun,return_sequences=False)(in0)
x = Dropout(0.1)(x)
x = Dense(units=50,activation=act_fun)(x)
x = Dropout(0.1)(x)
x = Dense(units=20,activation=act_fun)(x)
x = Dropout(0.1)(x)
x = Dense(units=5,activation=act_fun)(x)
x = Dropout(0.1)(x)
out0 = Dense(units=ytrain_oh.shape[1],activation='softmax')(x)

model = Model(in0,out0)
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['categorical_accuracy'])
```

```python
model.fit(Xtrain,ytrain_oh,batch_size=3,epochs=2,validation_split=0.1)
```
